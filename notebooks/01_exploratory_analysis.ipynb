```python
spreadspoke = pd.read_csv("../data/raw/spreadspoke_scores.csv")
spreadspoke.head()
spreadspoke.info()
spreadspoke.isna().sum()
```

```python
spreadspoke.columns = (
    spreadspoke.columns
    .str.strip()
    .str.lower()
    .str.replace(" ", "_")
)
```

```python
spreadspoke = spreadspoke.dropna(
    subset=["score_away","score_home","spread_favorite","over_under_line","team_favorite_id"]
)
```

```python
spreadspoke["weather_wind_mph"] = spreadspoke["weather_wind_mph"].fillna(spreadspoke["weather_wind_mph"].median())
spreadspoke["weather_humidity"] = spreadspoke["weather_humidity"].fillna(spreadspoke["weather_humidity"].median())
spreadspoke["weather_temperature"] = spreadspoke["weather_temperature"].fillna(spreadspoke["weather_temperature"].median())
spreadspoke["weather_detail"] = spreadspoke["weather_detail"].fillna("Unknown")
```

```python
spreadspoke["schedule_date"] = pd.to_datetime(spreadspoke["schedule_date"], errors="coerce")
```

```python
spreadspoke.to_csv("../data/processed/spreadspoke_scores_cleaned.csv")
```

```python
import sqlite3
import pandas as pd

conn= sqlite3.connect("../data/processed/nfl_betting.db")
```

```python
spreadspoke.to_sql(
    "spreadspoke_scores_cleaned", 
    conn, 
    if_exists="replace"
)
```

```python
ssc = """
    SELECT *
    FROM spreadspoke_scores_cleaned
    ORDER BY schedule_season DESC
    """

cleaned_spreadspoke = pd.read_sql(ssc, conn)
cleaned_spreadspoke
```

```python
nfl_sports_betting_final = pd.read_csv("../data/processed/spreadspoke_scores_cleaned.csv")

nfl_sports_betting_final.head()
nfl_sports_betting_final.info()
nfl_sports_betting_final.isna().sum()
```

```python
nfl_sports_betting_final["margin_of_victory"] = (
    nfl_sports_betting_final["score_home"] - nfl_sports_betting_final["score_away"]
)
```

```python
nfl_sports_betting_final["total_points"] = (
    nfl_sports_betting_final["score_home"] + nfl_sports_betting_final["score_away"]
)
```

```python
nfl_sports_betting_final["favorite_is_home"] = (
    nfl_sports_betting_final["team_favorite_id"] == nfl_sports_betting_final["team_home"]
)
```

```python
nfl_sports_betting_final["favorite_score"] = nfl_sports_betting_final.apply(
    lambda row: row["score_home"] if row["favorite_is_home"] else row["score_away"],
    axis=1
)
```

```python
nfl_sports_betting_final["underdog_score"] = nfl_sports_betting_final.apply(
    lambda row: row["score_away"] if row["favorite_is_home"] else row["score_home"],
    axis=1
)
```

```python
nfl_sports_betting_final["favorite_covered"] = (
    (nfl_sports_betting_final["favorite_score"] - nfl_sports_betting_final["underdog_score"]) >
    abs(nfl_sports_betting_final["spread_favorite"])
)
```

```python
nfl_sports_betting_final["over_under_line"] = pd.to_numeric(
    nfl_sports_betting_final["over_under_line"],
    errors="coerce"
)
```

```python
nfl_sports_betting_final["over_hit"] = (
    nfl_sports_betting_final["total_points"] >
    nfl_sports_betting_final["over_under_line"]
)
```

```python
nfl_sports_betting_final["over_under_line"] = nfl_sports_betting_final["over_under_line"].fillna(nfl_sports_betting_final["over_under_line"].median())
```

```python
nfl_sports_betting_final["spread_result"] = (
    (nfl_sports_betting_final["favorite_score"] - nfl_sports_betting_final["underdog_score"])
    - abs(nfl_sports_betting_final["spread_favorite"])
)
```

```python
nfl_sports_betting_final= nfl_sports_betting_final.drop(columns=["Unnamed: 0"])
```

```python
nfl_sports_betting_final.to_csv(
    "../data/processed/nfl_sports_betting_final.csv",
    index=False
)
```
